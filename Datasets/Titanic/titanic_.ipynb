Mostrar por pantalla el porcentaje de personas que sobrevivieron y murieron.

```python
titanic_df["Survived"].value_counts(normalize=True) * 100
```

```python
deaths_percent = round(np.sum(titanic_df["Survived"] == 0) / len(titanic_df["Survived"]) * 100, 2)
survivors_percent = round(np.sum(titanic_df["Survived"] == 1) / len(titanic_df["Survived"]) * 100, 2)
print("Survivors: ", survivors_percent, "\n" + "Deaths: ", deaths_percent)
```

```python
grades = ["Deaths", "Survivors"]
count = [deaths_percent, survivors_percent]
plt.bar(grades, count, width=0.4)
plt.xlabel("Grades")
plt.ylabel("Percentage")
for i in range(len(grades)):
    plt.text(i,count[i], str(count[i]), ha='center', va='bottom')
plt.show()
```

Eliminar del DataFrame los pasajeros con edad desconocida.

```python
titanic_df.dropna(subset=["Age"])
```

Mostrar por pantalla la edad media de las mujeres que viajaban en cada clase.

```python
mean_female = {}
for i in np.sort(titanic_df["Pclass"].unique()):
    j = titanic_df.loc[(titanic_df["Pclass"] == i) & (titanic_df["Sex"] == "female")]
    mean_female["Class " + str(i)] = round(np.mean(j["Age"]), 2)

mean_female
```

```python
print(titanic_df.groupby(['Pclass','Sex'])['Age'].mean().unstack()['female'])
```

Añadir una nueva columna booleana para ver si el pasajero era menor de edad o no.

```python
titanic_df["Younger"] = titanic_df["Age"] < 18
titanic_df
```

Mostrar por pantalla el porcentaje de menores y mayores de edad que sobrevivieron en cada clase.

```python
print(titanic_df.groupby(["Pclass", "Younger"])["Survived"].value_counts(normalize=True) * 100)
```
