```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime as dt
```

```python
em_2016 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/EmissionOfMadrid/emisiones-2016.csv", sep = ';')
em_2017 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/EmissionOfMadrid/emisiones-2017.csv", sep = ';')
em_2018 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/EmissionOfMadrid/emisiones-2018.csv", sep = ';')
em_2019 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Datasets/EmissionOfMadrid/emisiones-2019.csv", sep = ';')
```

Generar un DataFrame con los datos de los cuatro ficheros.

```python
em_concat = pd.concat([em_2016, em_2017, em_2018, em_2019])
em_concat.head(10)
```

Filtrar las columnas del DataFrame para quedarse con las columnas ESTACION, MAGNITUD, AÑO, MES y las correspondientes a los días D01, D02, etc.

```python
emissions = em_concat.drop(["PUNTO_MUESTREO", "PROVINCIA", "MUNICIPIO"], axis=1)
emissions
```

Reestructurar el DataFrame para que los valores de los contaminantes de las columnas de los días aparezcan en una única columna.

```python
emissions = emissions.melt(id_vars=["ESTACION", "MAGNITUD", "ANO", "MES"], var_name="DIA", value_name="EMISION")
emissions
```

Añadir una columna con la fecha a partir de la concatenación del año, el mes y el día (usar el módulo datetime).

```python
emissions["DIA"] = emissions["DIA"].str.strip("D")
emissions["FECHA"] = emissions["ANO"].apply(str) + "/" + emissions["MES"].apply(str) + "/" + emissions["DIA"].apply(str)
emissions["FECHA"] = pd.to_datetime(emissions["FECHA"], format="%Y/%m/%d", infer_datetime_format=True, errors="coerce")
emissions
```

Eliminar las filas con fechas no válidas (utilizar la función isnat del módulo numpy) y ordenar el DataFrame por estaciones contaminantes y fecha.

```python
emissions = emissions.drop(emissions[np.isnat(emissions["FECHA"])].index)
emissions.sort_values(['ESTACION', 'MAGNITUD', 'FECHA'])
```

Mostrar por pantalla las estaciones y los contaminantes disponibles en el DataFrame.

```python
print("Seasons: ", emissions["ESTACION"].unique())
print("Contaminants: ", emissions["MAGNITUD"].unique())
```

Crear una función que reciba una estación, un contaminante y un rango de fechas y devuelva una serie con las emisiones del contaminante dado en la estación y rango de fechas dado.

```python
def evolution(season, magnitud_cont, from_date, to_date):
    return emissions[(emissions["ESTACION"] == season) & (emissions.MAGNITUD == magnitud_cont) & (emissions.FECHA >= from_date) & (emissions.FECHA <= to_date)].sort_values("FECHA")["EMISION"]
evolution(56, 8, dt.datetime.strptime('2018/10/25', '%Y/%m/%d'), dt.datetime.strptime('2019/02/12', '%Y/%m/%d'))
```
